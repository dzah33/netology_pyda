### 6.Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test

#### 1) Нулевая гипотеза: расход топлива в городе (city-mpg) одинаковый у машин, использующих разный тип топлива (fuel-type)

```python
gas_city = df[df['fuel-type']=='gas']['city-mpg']
diesel_city = df[df['fuel-type']=='diesel']['city-mpg']
```

```python
len(gas_city), len(diesel_city)
```

```python
gas_city = gas_city.sample(20)
diesel_city = diesel_city.sample(20)
```

```python
statistics.variance(gas_city), statistics.variance(diesel_city)
```

```python
sts.ttest_ind(gas_city, diesel_city, equal_var=False)
```

###### полученное значение pvalue=0.028, что меньше 0,05. Таким образом мы можем отклонить нулевую гипотезу об одинаковом расходе топлива в городе у автомобилей с бензиновым и дизельным двигателем

#### 2) Нулевая гипотеза: расход топлива на трассе (highway-mpg) одинаковый у автомобилей с бензиновым и дизельным двигаетелем (fuel-type)

```python
gas_highway = df[df['fuel-type']=='gas']['highway-mpg']
diesel_highway = df[df['fuel-type']=='diesel']['highway-mpg']
```

```python
len(gas_highway), len(diesel_highway)
```

```python
gas_highway = gas_highway.sample(20)
diesel_highway = diesel_highway.sample(20)
```

```python
statistics.variance(gas_highway), statistics.variance(diesel_highway)
```

```python
sts.ttest_ind(gas_highway, diesel_highway, equal_var=False)
```

#### полученное значение pvalue=0.0948, что больше 0,05. Таким образом мы не можем отклонить нулевую гипотезу
