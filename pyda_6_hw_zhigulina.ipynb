### Задание 2.
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
# смотрим правильное наименование стран
power['country'].unique()
```

```python
# фильтруем страны Прибалтики
baltic = power[ (power['country']=='Latvia') | (power['country']=='Lithuania') |(power['country']=='Estonia') ]
baltic.head()
```

```python
# фильтруем по категории и отбрасываем отрицательные значения quantity
bal_cat = baltic[ ((baltic['category'] == 4) | (baltic['category'] == 12) | (baltic['category'] == 21)) & (baltic['quantity'] >= 0)]

# фильтруем по году
bal_cat = bal_cat[ (bal_cat['year'] >= 2005) & (bal_cat['year'] <= 2010) ]

# считаем суммарное потребление

print(f'Cуммарное потребление стран Прибалтики c 2005 по 2010 год: {bal_cat.quantity.sum()}')
```
